## DATA MANIPULATION TECHNIQUES

### Insert a column, row, cell

```python
# Insert a column
data.insert(10,"Govt Exams","Yes")
```

```python
data.head()
```

```python
# Insert a Row
df={
    "Roll No":["20BCE0049"],
    "Name":["Samridh"],
    "Age" :[20],
    "DOB":["10-12-2002"],
    "CGPA":[9.95],
    "Courses":[10],
    "Graduation Year":[2024],
    "Placements":["Yes"],
    "M.Tech/MS":["Yes"],
    "Startup":["No"],
    "Govt Exams":["Yes"]
}
new_row = pd.DataFrame(df)
new_row.to_csv('Untiled.csv', mode='a', index=False, header=False)
data = data.append(new_row)
```

```python
# Insert a Cell
data.iloc[0,1]="Sahil"
```

```python
data
```

### Delete a Column, Row, Cell

```python
# Delete a Column
data = data.drop("Govt Exams",axis=1).head()
```

```python
# Delete a Row
data = data.drop(0,axis=0).head()
```

```python
# Delete a Cell
data.iloc[1,2]="NaN"
```

```python
data.head()
```
